```python
from modules import read_from_allfiles
import re

docs = read_from_allfiles.doc_from_allfiles()
alltexts = ''.join(docs)
abbs = []
matches = re.finditer(r'\(([a-zA-Z\u0391-\u03c9][a-zA-Z\d\-–\/\u2215]+)\)', alltexts) #greek = re.compile(r'[\u0391-\u03c9]') # ギリシャ文字
# matches = re.finditer(r'\(([a-zA-Z\u0391-\u03c9][^\s\^\{\}\.\,\;\:\(\)\[\]]+?)s?\)', alltexts)
for match in matches:
    if not re.match(r'[a-z]+', match.group(1)):
        abbs.append(match.group(1))

abb_list = list(set(abbs)) 
```

```python
### using reg expression
words = abb_list[0:]

new_ph_dic = {}
for word in tqdm(words):
    matches = re.finditer(fr'\(({word})\)', alltexts)
    new_phrase_list = []
    for m in matches:  
        wis = m.start()
        i = wis
        dis = wis
        while i > wis-80:
            if re.match(r'[\.\,\;\:\)\[\]]' ,alltexts[i:i+1]):
                dis = i+1
                break
            else:
                dis = i
            i -= 1
        
        ptext = (alltexts[dis:m.end()])
        
        char_set = []
        for char in m.group(1).lower().replace('-', ''):
            char_set.append(char)
        term1 = '(.*?)'.join(char_set) + '(.*?)'
        abb_reg1 = re.compile(fr'{term1}\({word}\)', re.I) # loose match
        term2 = '(.*?)[\s\-]'.join(char_set) + '(.*?)'
        abb_reg2 = re.compile(fr'{term2}\({word}\)', re.I) # tight match
        
        wl = ptext.split(' ')
        j = len(wl)-1
        new_phrase = ''
        while j > 0: # tight matching loop
            stext = ' '.join(wl[j:])
            if re.match(abb_reg2, stext):
                new_phrase = stext
                break
            j -= 1
        j = len(wl)-1
        if new_phrase == '': # loose matching loop
            while j > 0:
                stext = ' '.join(wl[j:])
                if re.match(abb_reg1, stext):
                    new_phrase = stext
                    break
                j -= 1
            
        if new_phrase:
            new_phrase_list.append(new_phrase)
        
    new_ph_dic[word] = set(new_phrase_list)
```
